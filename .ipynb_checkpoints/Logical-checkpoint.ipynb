```python
from kanren import *

tools = Relation()
facts(tools, ('Language', 'Python'), 
             ('Language', 'Java'),
             ('Package', 'Word'),
             ('Package', 'Excel'),
             ('OS', 'Windos'),
             ('Os', 'Linux'),
             ('CS', 'Language'),
             ('CS', 'OS'),
             ('Non CS', 'Packages'))

class_v, tools_v = var(), var()
print(run(0, class_v, tools(class_v, tools_v), tools (tools_v, 'Java')))

```

```python
from kanren import *

tools = Relation()
courses = Relation()
facts(courses, ('Language', 'Python'), 
             ('Language', 'Java'),
             ('Package', 'Word'),
             ('Package', 'Excel'),
             ('Os', 'Windows'),
             ('Os', 'Linux'))

facts(tools,('OS', 'CS'),
             ('Languages', 'CS'),
             ('Packages', 'Non Cs'))

class_v, tools_v = var(), var()
print(run(0, tools_v, tools(class_v, tools_v), courses(class_v, 'Non Cs')))

```
