```python
from kanren import *
from kanren.constraints import *

p, q, r = var(), var(), var()

# modus ponen
print(run(0, q, eq(p, q), membero(p, (True, False)), membero(q, (True, False)), eq(p, True)))

# modus tollens
print(run(0, q, eq(p, q), membero(p, (True, False)), membero(q, (True, False)), eq(p, False)))

# hypothetical syllogysn
print(run(0, p, eq(p, q), eq(q, r), membero(p, (True, False)), membero(q, (True, False)), membero(r, (True, False)), eq(r, True)))

# disjunctive syllogysm
print(run(0, (p,q), lany(eq(p, True), eq(q, True)), membero(p, (True, False)), membero(q, (True, False)), neq(p, True)))

# Addition rule
print(set(run(0, (p,q), lany(eq(p, True), eq(q, True)), membero(p, (True, False)), membero(q, (True, False)), eq(p, True))))

# Simplification rule
print(run(0, (p,q), lall(eq(p, True), eq(q, True)), membero(p, (True, False)), membero(q, (True, False)), eq(p, True)))
```

```python
symptoms = Relation()
diagnosis = Relation()
facts(symptoms, ('Flu', 'Fever'), ('Flu', 'Cough'), ('Flu', 'Runny Nose'), ('Flu', 'Headache'),
      ('Cold', 'Cough'), ('Cold', 'Runny Nose'), ('Cold', 'Headache'))
facts(diagnosis, ('Flu', 'Antibiotics'), ('Cold', 'Antihistamine'))

disease_v, symptoms_v = var(), var()
print(set(run(0, disease_v, diagnosis(disease_v, symptoms_v), diagnosis(disease_v, 'Antibiotics'))))
print(set(run(0, disease_v, symptoms(disease_v, symptoms_v), symptoms(disease_v, 'Fever'))))
print(set(run(0, disease_v, symptoms(disease_v, symptoms_v), symptoms(disease_v, 'Headache'))))
```

```python
#redKnight redKnave blueKnight blue Knave
#T           F         F          T
#F           T         T          F

from kanren import *
from kanren.constraints import *

redKnight, redKnave, blueKnight, blueKnave = var(), var(), var(), var()

rules = lany(lall(eq(redKnight, 'T'), eq(redKnave, 'F'), eq(blueKnight, 'F'),eq(blueKnave, 'T')),
             lall(eq(redKnight, 'F'), eq( redKnave, 'T'), eq(blueKnight, 'T'),eq(blueKnave, 'F')))

print('Red says I am both a Knight and a Knave')

if(bool(run(0, (redKnight, redKnave, blueKnight, blueKnave), rules, eq(redKnight, 'T'), eq(redKnave, 'T'))) == True):
    print('You are a Knight')
else:
    print('You are a Knave')
```

```python
adjacent = Relation()
coastal = Relation()
coastalStates = 'NY, FL , CA, WA, TX, HW'.split(',')
for state in coastalStates:
    fact(coastal, state)
with open('AdjacentStates.txt') as f:
    adjacentList = [line.strip(). split(',') for line in f
                   if line and line[0]. isalpha()]
for L in adjacentList:
    head, tail = L[0], L[1:]
    for state in tail:
        fact(adjacent, head, state)
        
x, y, z = var(), var(), var('Yes')
print('Is NY a coastal state? \n')
print (run(0, z, coastal('NY')))

print('Is CA adjacent to NY? \N')
if(bool(run(0, x, adjacent('CA', 'NY'))) == True):
    print('Yes')
else:
    print('No')
```
