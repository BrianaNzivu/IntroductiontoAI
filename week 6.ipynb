```python
#depth first search
import random,operator
class Environment(object):
    myGraph={"1":set(["2", "4"]),
            "2":set(["1","3", "5"]),
            "3":set(["2", "6"]),
            "4":set(["5", "7"]),
            "5":set(["4","6", "8"]),
            "6":set(["5", "9"]),
            "7":set(["4", "8"]),
            "8":set(["7","5", "9"]),
            "9":set(["6", "8"]),}
    cost={str(["1","2"]): "3", str(["1","4"]): "5",
         str(["2","1"]): "3", str(["2","5"]): "7", str(["2","3"]): "5",
         str(["3","2"]): "5", str(["3","6"]): "9",
         str(["4","1"]): "5", str(["4","5"]): "9", str(["4","7"]): "11",
         str(["5","2"]): "7", str(["5","4"]): "9", str(["5","8"]): "13", str(["5","6"]): "11",
         str(["6","3"]): "9", str(["6","9"]): "15", str(["6","5"]): "11",
         str(["7","4"]): "11", str(["7","8"]): "15",
         str(["8","5"]): "13", str(["8","7"]): "15", str(["8","9"]): "17",
         str(["9","6"]): "15", str(["9","8"]): "17",

         }
    myCoordinates={"1":["1","3"],"2":["2","3"],"3":["3","3"],
                   "4":["2","2"],"5":["2","2"],"6":["3","2"],
                   "7":["1","1"],"8":["2","1"],"9":["3","1"],
                  }
    state="1"
    goal="9"
    
class Agent(Environment):
    def dfs (self, graph, start, goal):
        stack=[(start,[start])]# current + journey
        p=[] # empty coz there's no journey
        while stack: #while the stack is not empty coz haven't reached destination
            (vertex, path)= stack.pop()# getting sth out of a stack
           # print("Vertex and path:", vertex, path)# path= previous location, vertex= current
           # print("Stack: ", stack)
            #exit while loop once the stack is empty
            for next in graph[vertex]- set(path):# neighbours of graph in position vertex except the ones visited before(path)
                if next==goal:# is it the destination
                    p.append(path+ [next])#adding current location to create a journey
                else:#if didn't reach final destination
                    stack.append((next, path+[next])) #journey + the current position
                    #print("Stack2: ", stack)
        return p # possible routes to the goal
    
    def bfs (self, graph, start, goal):
        stack=[(start,[start])]# current + journey
        p=[] # empty coz there's no journey
        while stack: #while the stack is not empty coz haven't reached destination
            (vertex, path)= stack.pop(0)# getting sth out of a stack
           # print("Vertex and path:", vertex, path)# path= previous location, vertex= current
            #print("Stack: ", stack)
            #exit while loop once the stack is empty
            for next in graph[vertex]- set(path):# neighbours of graph in position vertex except the ones visited before(path)
                if next==goal:# is it the destination
                    p.append(path+ [next])#adding current location to create a journey
                    return p
                else:#if didn't reach final destination
                    stack.append((next, path+[next])) #journey + the current position
                    #print("Stack2: ", stack)
        return p # possible routes to the goal
    def getcost(path_to_cost):
        i=0
        pathCost=0
        while (i<len(path_to_cost) - 1):
            l=[]
            l.append(path_to_cost[i])
            l.append(path_to_cost[i +1])
            pathCost=pathCost+ int(Environment.cost[str(l)])
            i+=1
        return pathCost
    
    def UCS(self, graph, start, goal):
        stack=[(start,[start])]# current + journey
        p=[] # empty coz there's no journey
        c=[]
        leastCost=1000
        while stack: #while the stack is not empty coz haven't reached destination
            (vertex, path)= stack.pop(0)# getting sth out of a stack
            # print("Vertex and path:", vertex, path)# path= previous location, vertex= current
            # print("Stack: ", stack)
            #exit while loop once the stack is empty
            for next in graph[vertex]- set(path):# neighbours of graph in position vertex except the ones visited before(path)
                if next==goal:# is it the destination
                    pathCost=Agent.getcost(path+[next])
                    print("Path", path+[next], "Path Cost", pathCost)
                    if pathCost < leastCost:
                        leastCost= pathCost
                        p= path + [next]
                else:#if didn't reach final destination
                    stack.append((next, path+[next])) #journey + the current position
                    #print("Stack2: ", stack)
        return p # possible routes to the goal

    
    def __init__(self, Environment):
        print("DFS: ", self.dfs(Environment.myGraph, Environment.state, Environment.goal))
        print("BFS: ", self.bfs(Environment.myGraph, Environment.state, Environment.goal))
        print("UCS: ", self.UCS(Environment.myGraph, Environment.state, Environment.goal))
        print("Gbfs: ", self.gbfs(Environment.myGraph, Environment.state, Environment.goal))
        print("A-star: ", self.Astar(Environment.myGraph, Environment.state, Environment.goal))

    def getheuristics(vertex, goal):
        v=[]
        g=[]
        for i in Environment.myCoordinates[vertex]:
            v.append(int(i))
            
        for i in Environment.myCoordinates[goal]:
            g.append(int(i))   
        m= abs(v[0]-g[0]) + abs(v[1]-g[1])
        return m
    
    def gbfs(self, graph, start, goal):
        p=[]
        p.append(start)
        while True:
            neighbour=graph[start]
            heuristics={}
            for i in neighbour.difference(p):#No visiting a position twice
                heuristics[i]=Agent.getheuristics(i, goal)
                
            sortedheuristics=sorted(heuristics.items(), key=operator.itemgetter(1))
            h=next(iter(sortedheuristics[0]))
            p.append(h)
            if h==goal:
                return p
            else:
                start=h
                
     
    def Astar(self, graph, start, goal):
        p=[]
        p.append(start)
        while True:
            neighbour=graph[start]
            heuristics={}
            for i in neighbour.difference(p):#No visiting a position twice
                l=[]
                l.append(str(start))
                l.append(str(i))
                heuristics[i]=Agent.getheuristics(i, goal) + Agent.getcost(l)
            sortedheuristics=sorted(heuristics.items(), key=operator.itemgetter(1))
            h=next(iter(sortedheuristics[0]))
            p.append(h)
            if h==goal:
                return p
            else:
                start=h
        
TheEnvironment= Environment()
TheAgent=Agent(TheEnvironment)
```
